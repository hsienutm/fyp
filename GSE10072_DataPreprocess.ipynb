**1. Use GEOparse to obtain the corresponding gene entrez id for the probe id of the dataset.**

```python
import GEOparse
gse1=GEOparse.get_GEO(geo="GSE10072", destdir="./")
gse1.gsm = gse1.phenotype_data
probeList1=gse1.gpls['GPL96'].table
probe1=probeList1['ID']
outfile=open('GSE10072_probe.txt','w')
for p1 in probe1:
    outfile.write(p1+'\n')
outfile.close()
```

**2. Import the required libraries to do data pre-processing.**

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

**3. Read the datasets and save them in dataframes.**

```python
df1 = pd.read_csv('D:\A_FYP\Dataset\GSE10072.csv') # The original gene expression data
df2 = pd.read_csv('D:\A_FYP\Dataset\Entrez_GSE10072.csv') # The conversion data from probe id to gene entrez id
```

**4. Replace the probe id in df1 with the respective gene entrez id in df2.**

```python
df1['ID'] = df1['ID'].map(df2.set_index('From')['To'])
```

**Show the first five rows to verify the replacement**

```python
df1.head()
```

**5. Use *inna()* method available in pandas to check for the total number of missing values.**

```python
pd.isna(df1['ID']).sum() #missing values 
```

**6. Use *dropna()* method to drop all the missing values from the dataframe.**

```python
df1.dropna(subset = ["ID"], inplace=True)
```

**7. Check whether or not there are still missing values.**

```python
pd.isna(df1['ID']).sum()
```

**8. Check for the duplicated gene entrez id in the dataframe by using *duplicated()* method.**

```python
df1['ID'].duplicated().sum()
```

**9. Average the duplicated values and replace it with the mean values. This is done by using the *mean()* function. Then save the sorted dataframe into a new dataframe, called df3.**

```python
df3 = df1.groupby(['ID'], sort=False).mean()
```
