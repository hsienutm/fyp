```python
#comparision between best and other cluster GSE10072
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.3
fig = plt.subplots(figsize =(12, 8))

# set height of bar
best = [90, 94.97, 85.05]
other = [100, 100, 100]


# Set position of bar on X axis
br1 = np.arange(len(best))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, best, color ='none', width = barWidth,
		edgecolor ='black', label ='Best Cluster',hatch='+')
plt.bar(br2, other, color ='none', width = barWidth,
		edgecolor ='black', label ='Other Clusters',hatch='x')

# Adding Xticks
plt.xlabel('GSE10072 Lung Dataset', fontweight ='bold', fontsize = 15)
plt.ylabel('Percentage (%)', fontweight ='bold', fontsize = 15)
plt.title('Comparision of performance between Best and Other clusters GSE10072')
plt.xticks([r +0.15 for r in range(3)],
		['Accuracy', 'Sensitivity', 'Specificity'])

plt.legend()
plt.show()
```
