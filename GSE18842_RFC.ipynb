```python
df = pd.read_csv('D:\A_FYP\Dataset/GSE18842_clean_transposed.csv')
df
```

```python
df_set = df.iloc[:,2:]
```

```python
target = df_set["class"]
predictor = df_set.drop("class", axis=1)
```

```python
lx_train, lx_test, ly_train, ly_test = train_test_split(predictor,target,test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score, train_test_split

rf = RandomForestClassifier()
rf.fit(lx_train,ly_train)
Y_pred_rf = rf.predict(lx_test)
accuracy = round(accuracy_score(ly_test,Y_pred_rf)*100,2)

cm = metrics.confusion_matrix(ly_test, Y_pred_rf)
 
# Showing the confusion matrix
#current_sensivity, current_specificty = confusion_metrics(cm_df)
TN = cm[1][1]

TP = cm[0][0]

FN = cm[0][1]

FP = cm[1][0]
conf_sensitivity = (TP / float(TP + FN))
conf_specificity = (TN / float(TN + FP))
sensivity =round(conf_sensitivity*100,2)
specificty = round(conf_specificity*100,2) 

print("accuracy: ", accuracy)
print("sensivity: ", sensivity)
print("specificty: ", specificty)
```
